```python
import tensorflow as tf
from tensorflow.keras.datasets import cifar10

# Load the CIFAR-10 dataset
(x_train, _), (x_test, _) = cifar10.load_data()

# Normalize the images to the range [0, 1]
x_train = x_train.astype('float32') / 255.0
x_test = x_test.astype('float32') / 255.0

# Set the input shape for RGB images
input_shape = (32, 32, 3)
```

```python
from tensorflow.keras import layers, models

def build_cifar10_autoencoder_rgb(input_shape):
    # Encoder
    encoder_input = layers.Input(shape=input_shape)
    x = layers.Conv2D(32, (3, 3), activation='relu', padding='same')(encoder_input)
    x = layers.MaxPooling2D((2, 2), padding='same')(x)
    x = layers.Conv2D(64, (3, 3), activation='relu', padding='same')(x)
    x = layers.MaxPooling2D((2, 2), padding='same')(x)

    # Latent space
    x = layers.Flatten()(x)
    latent = layers.Dense(128, activation='relu')(x)

    # Decoder
    x = layers.Dense(8*8*64, activation='relu')(latent)
    x = layers.Reshape((8, 8, 64))(x)
    x = layers.Conv2DTranspose(64, (3, 3), activation='relu', padding='same')(x)
    x = layers.UpSampling2D((2, 2))(x)
    x = layers.Conv2DTranspose(32, (3, 3), activation='relu', padding='same')(x)
    x = layers.UpSampling2D((2, 2))(x)
    decoder_output = layers.Conv2D(3, (3, 3), activation='sigmoid', padding='same')(x)

    # Autoencoder Model
    autoencoder = models.Model(encoder_input, decoder_output)
    autoencoder.compile(optimizer='adam', loss='mse')
    return autoencoder

# Build and summarize the model
autoencoder_rgb = build_cifar10_autoencoder_rgb(input_shape)
autoencoder_rgb.summary()
```

```python
# Train the autoencoder on RGB images
history = autoencoder_rgb.fit(
    x_train, x_train,
    epochs=10,
    batch_size=64,
    validation_split=0.1
)
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np

# Predict reconstructed images for the test set
reconstructed_images = autoencoder_rgb.predict(x_test)

# Calculate reconstruction errors for each test image
recon_errors = []
for i in range(len(x_test)):
    original = x_test[i].reshape(-1)
    reconstructed = reconstructed_images[i].reshape(-1)
    error = mean_squared_error(original, reconstructed)
    recon_errors.append(error)

# Calculate average reconstruction error
avg_recon_error = np.mean(recon_errors)
std_recon_error = np.std(recon_errors)

print(f"Average Reconstruction Error (MSE): {avg_recon_error:.4f}")
print(f"Standard Deviation of Reconstruction Error: {std_recon_error:.4f}")
```

# Anomaly Detection / Adversarial Prediction
### Normal Image

```python
import numpy as np
from sklearn.metrics import mean_squared_error

# Generate reconstructed images for the test set
reconstructed_images = autoencoder_rgb.predict(x_test)

# Calculate reconstruction errors for each test image
recon_errors = []
for i in range(len(x_test)):
    original = x_test[i].reshape(-1)
    reconstructed = reconstructed_images[i].reshape(-1)
    error = mean_squared_error(original, reconstructed)
    recon_errors.append(error)

# Calculate the mean and standard deviation of reconstruction errors on normal data
avg_recon_error = np.mean(recon_errors)
std_recon_error = np.std(recon_errors)

# Set a threshold for anomaly detection (e.g., mean + 2 * standard deviation)
anomaly_threshold = avg_recon_error + 2 * std_recon_error

# Print the threshold and average reconstruction error
print(f"Average Reconstruction Error (MSE): {avg_recon_error:.4f}")
print(f"Threshold for Anomaly Detection: {anomaly_threshold:.4f}")

# Function to detect anomalies/adversarial images based on the threshold
def detect_anomaly(image):
    reconstructed = autoencoder_rgb.predict(image.reshape(1, *image.shape)).reshape(-1)
    original = image.reshape(-1)
    error = mean_squared_error(original, reconstructed)
    return error, error > anomaly_threshold

sample_image = x_test[0]  # Replace with any test image
error, is_anomalous = detect_anomaly(sample_image)
print(f"Reconstruction Error: {error:.4f}")
print("Anomalous" if is_anomalous else "Normal")
```

### Noise-Induced Image

```python
import cv2
# Add noise to a test image (e.g., Gaussian noise)
def add_noise(image, noise_factor=0.25):
    noisy_image = image + noise_factor * np.random.normal(loc=0.0, scale=1.0, size=image.shape)
    noisy_image = np.clip(noisy_image, 0., 1.)  # Keep pixel values in the range [0, 1]
    return noisy_image

# Generate a noisy version of the sample image
sample_image = x_test[0]  # Replace with any test image
noisy_sample_image = add_noise(sample_image)

# Detect anomaly using the noisy image
error, is_anomalous = detect_anomaly(noisy_sample_image)
print(f"Reconstruction Error (Noisy Image): {error:.4f}")
print("Anomalous" if is_anomalous else "Normal")


# Apply Gaussian blur to an image
def add_blur(image, kernel_size=(5, 5)):
    blurred_image = cv2.GaussianBlur(image, kernel_size, 0)
    return blurred_image

# Generate a blurred version of the sample image
sample_image = x_test[0]  # Replace with any test image
blurred_sample_image = add_blur(sample_image)

# Detect anomaly using the blurred image
error, is_anomalous = detect_anomaly(blurred_sample_image)
print(f"Reconstruction Error (Blurred Image): {error:.4f}")
print("Anomalous" if is_anomalous else "Normal")

# Plotting the original, noisy, and blurred images
fig, axs = plt.subplots(1, 3, figsize=(12, 4))

# Display the original image
axs[0].imshow(sample_image)
axs[0].set_title("Original Image")
axs[0].axis("off")

# Display the noisy image
axs[1].imshow(noisy_sample_image)
axs[1].set_title("Noisy Image")
axs[1].axis("off")

# Display the blurred image
axs[2].imshow(blurred_sample_image)
axs[2].set_title("Blurred Image")
axs[2].axis("off")

plt.show()
```

```python
# Adjust hue of the image
def adjust_hue(image, hue_shift=30):
    hsv_image = cv2.cvtColor((image * 255).astype(np.uint8), cv2.COLOR_RGB2HSV)
    hsv_image[..., 0] = (hsv_image[..., 0] + hue_shift) % 180  # Shift hue channel
    return cv2.cvtColor(hsv_image, cv2.COLOR_HSV2RGB) / 255.0  # Convert back to RGB and normalize

# Increase saturation of the image
def increase_saturation(image, saturation_factor=2.0):
    hsv_image = cv2.cvtColor((image * 255).astype(np.uint8), cv2.COLOR_RGB2HSV)
    hsv_image[..., 1] = np.clip(hsv_image[..., 1] * saturation_factor, 0, 255)  # Increase saturation
    return cv2.cvtColor(hsv_image, cv2.COLOR_HSV2RGB) / 255.0  # Convert back to RGB and normalize

# Generate hue-shifted and saturation-enhanced versions of the sample image
sample_image = x_test[1]  # Replace with any test image
hue_shifted_image = adjust_hue(sample_image, hue_shift=30)
saturated_image = increase_saturation(sample_image, saturation_factor=2.0)

# Plot the original, hue-shifted, and saturation-enhanced images
fig, axs = plt.subplots(1, 3, figsize=(12, 4))

# Display the original image
axs[0].imshow(sample_image)
axs[0].set_title("Original Image")
axs[0].axis("off")

# Display the hue-shifted image
axs[1].imshow(hue_shifted_image)
axs[1].set_title("Hue-Shifted Image")
axs[1].axis("off")

# Display the saturation-enhanced image
axs[2].imshow(saturated_image)
axs[2].set_title("Saturation-Enhanced Image")
axs[2].axis("off")

plt.show()
# Detect anomalies on hue-shifted and saturation-enhanced images
error_hue, is_anomalous_hue = detect_anomaly(hue_shifted_image)
error_saturation, is_anomalous_saturation = detect_anomaly(saturated_image)

# Display the results
print("Anomaly Detection Results:")
print(f"Hue-Shifted Image - Reconstruction Error: {error_hue:.4f} - {'Anomalous' if is_anomalous_hue else 'Normal'}")
print(f"Saturation-Enhanced Image - Reconstruction Error: {error_saturation:.4f} - {'Anomalous' if is_anomalous_saturation else 'Normal'}")
```

```python
import tensorflow.keras.backend as K

def semantic_loss(original, reconstructed):
    # Extract features for original and reconstructed images
    original_features = feature_extractor(original)
    reconstructed_features = feature_extractor(reconstructed)
    
    # Compute normalized L2 distance in the feature space
    loss = K.mean(K.square(original_features - reconstructed_features))
    return loss

def weighted_semantic_loss(original, reconstructed, weight=1.0):
    feature_loss = semantic_loss(original, reconstructed)
    return feature_loss * weight
```

```python
from tensorflow.keras.optimizers.schedules import ExponentialDecay

initial_lr = 1e-3
lr_schedule = ExponentialDecay(
    initial_lr, decay_steps=10000, decay_rate=0.9, staircase=True)
optimizer = tf.keras.optimizers.Adam(learning_rate=lr_schedule)
autoencoder_rgb.compile(optimizer=optimizer, loss=weighted_semantic_loss)

# Train the autoencoder
autoencoder_rgb.fit(x_train, x_train, epochs=50, batch_size=64, validation_split=0.1)
```
